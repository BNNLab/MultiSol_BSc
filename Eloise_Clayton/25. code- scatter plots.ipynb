```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load data
data = pd.read_csv('MLR_predictions.csv')

# Extract the necessary columns
predicted = data['Predicted']
experimental = data['Experimental']

# Create the scatter plot
plt.figure(figsize=(10, 10))

plt.scatter(experimental, predicted, color='black', s=15)

x_vals = np.linspace(-8, 2, 100)
plt.plot(x_vals, x_vals, 'k', linewidth=0.8)
plt.plot(x_vals, x_vals + 1, 'r', linewidth=0.8)
plt.plot(x_vals, x_vals - 1, 'r', linewidth=0.8)
plt.plot(x_vals, x_vals + 0.7, 'b', linewidth=0.8)
plt.plot(x_vals, x_vals - 0.7, 'b', linewidth=0.8)

# Axis limits
plt.xlim(-5, 2)
plt.ylim(-5, 2)

# Labels and title
plt.title('Experimental vs Predicted\nLog S using MLR', fontsize=30, pad=30)
plt.xlabel('Experimental Log S', fontsize=26)
plt.ylabel('Predicted Log S', fontsize=26)

plt.text(-4.7, 1.6, "% within 1 = 81.10", fontsize=26, color='red')
plt.text(-4.7, 1.2, "% within 0.7 = 63.33", fontsize=26, color='blue')

# No grid
plt.grid(False)

plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
           
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load data
data = pd.read_csv('MLR_predictions.csv')

# Extract the necessary columns
predicted = data['Predicted']
experimental = data['Experimental']
absolute_error = data['Absolute Error']  

# Define threshold for acceptable error
threshold = 1.0

# Calculate how many points have absolute error less than the threshold
within_error_range = absolute_error < threshold
percent_within = within_error_range.sum() / len(data) * 100

# Create the scatter plot
plt.figure(figsize=(10, 10))

# Smaller dots (s=15 makes them smaller than default ~36)
plt.scatter(experimental, predicted, color='black', s=15)

# Thinner lines (linewidth=0.8 for example)
x_vals = np.linspace(-25, 2, 100)
plt.plot(x_vals, x_vals, 'k', linewidth=0.8)
plt.plot(x_vals, x_vals + 1, 'r', linewidth=0.8)
plt.plot(x_vals, x_vals - 1, 'r', linewidth=0.8)
plt.plot(x_vals, x_vals + 0.7, 'b', linewidth=0.8)
plt.plot(x_vals, x_vals - 0.7, 'b', linewidth=0.8)

# Axis limits
plt.xlim(-25, 2)
plt.ylim(-5, 2)

# Labels and title
plt.title('Experimental vs Predicted\nLog S using MLR', fontsize=30, pad=30)
plt.xlabel('Experimental Log S', fontsize=26)
plt.ylabel('Predicted Log S', fontsize=26)

plt.text(-24, 1.5, "% within 1 = 81.10", fontsize=26, color='red')
plt.text(-24, 1.1, "% within 0.7 = 63.33", fontsize=26, color='blue')

# No grid
plt.grid(False)

plt.xticks(fontsize=18)
plt.yticks(fontsize=18) 

plt.tight_layout()
plt.show()
```
