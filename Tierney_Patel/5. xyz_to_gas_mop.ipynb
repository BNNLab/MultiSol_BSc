```python
import os

# Define the path to the xyz files
xyz_dir = "/Users/stella/Documents/tierney/project/notebooks/xyz"

# Define the path to the mop_gas directory (in the same parent directory as xyz_files)
parent_dir = os.path.dirname(xyz_dir)  # Get the parent directory of xyz_files
mop_gas_dir = os.path.join(parent_dir, 'gas_mopac_files')  # Create mop_gas in the same parent directory

# Create the mop_gas directory if it doesn't exist
if not os.path.exists(mop_gas_dir):
    os.makedirs(mop_gas_dir)
    print(f"Created directory: {mop_gas_dir}")

# Process each .xyz file in the directory
for filename in os.listdir(xyz_dir):
    if filename.endswith('.xyz') and filename.startswith('PM6_'):
        # Extract the solute_inchikey from the filename
        solute_inchikey_with_suffix = filename.replace('PM6_', '').replace('.mop', '')
        solute_inchikey = solute_inchikey_with_suffix.split('_')[0]  # Remove the suffix
        

        # Generate the corresponding .mop filename
        mop_filename = filename.replace('.xyz', '.mop')
        full_mop_filename = os.path.join(mop_gas_dir, mop_filename)  # Save in mop_gas folder
        full_xyz_filename = os.path.join(xyz_dir, filename)  # Full path to the .xyz file

        # Open the .xyz file and read its content
        with open(full_xyz_filename, 'r') as xyz_file:
            lines = xyz_file.readlines()[2:]  # Skip the first two lines (atom count and comment)

        # Write the .mop file
        with open(full_mop_filename, 'w') as mop_file:
            mop_file.write('AUX LARGE OPT FORCE THERMO PM6 T=128H RECALC=5 GNORM=0.01 LET SCFCRT=0.0000001\n')
            mop_file.write(f"{solute_inchikey}\n")  # Add the solute_inchikey as the second command line
            mop_file.writelines(lines)  # Write the atomic coordinates

        print(f"Generated MOPAC file: {full_mop_filename}")

print("All .xyz files have been processed.")
```
